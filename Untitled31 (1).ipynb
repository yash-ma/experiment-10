```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# ------------------------------------------------
# ✅ Task 1: Define GridWorld (States, Rewards, T)
# ------------------------------------------------

ROWS, COLS = 3, 4

states = []
walls = [(1, 1)]
terminal_states = [(0, 3), (1, 3)]

for r in range(ROWS):
    for c in range(COLS):
        if (r, c) not in walls:
            states.append((r, c))

ACTIONS = ['up', 'down', 'left', 'right']

# Default rewards
R = {
    (0, 3): +1,
    (1, 3): -1
}

living_penalty = -0.04

def get_reward(state):
    if state in R:
        return R[state]
    return living_penalty

gamma = 0.99

# ------------------------------------------------
# Helper to compute valid transitions
# ------------------------------------------------
def move(state, action):
    r, c = state
    if action == 'up': r -= 1
    elif action == 'down': r += 1
    elif action == 'left': c -= 1
    elif action == 'right': c += 1

    if (r < 0 or r >= ROWS or c < 0 or c >= COLS or (r, c) in walls):
        return state
    return (r, c)

# ------------------------------------------------
# T(s, a) → list of (probability, next_state)
# ------------------------------------------------
def get_next_states(state, action):
    if state in terminal_states:
        return [(1.0, state)]

    # Intended and slip directions
    intended = action
    if action == 'up': left, right = 'left', 'right'
    if action == 'down': left, right = 'right', 'left'
    if action == 'left': left, right = 'down', 'up'
    if action == 'right': left, right = 'up', 'down'

    transitions = []
    transitions.append((0.8, move(state, intended)))
    transitions.append((0.1, move(state, left)))
    transitions.append((0.1, move(state, right)))
    return transitions
```

```python
# ------------------------------------------------
# ✅ Task 2: VALUE ITERATION
# ------------------------------------------------
def value_iteration(theta=0.0001):
    V = {s: 0.0 for s in states}
    iteration = 0

    while True:
        delta = 0
        new_V = V.copy()

        for s in states:
            if s in terminal_states:
                new_V[s] = 0
                continue

            action_values = []
            for a in ACTIONS:
                q_sa = 0
                for prob, next_s in get_next_states(s, a):
                    q_sa += prob * (get_reward(next_s) + gamma * V[next_s])
                action_values.append(q_sa)

            new_V[s] = max(action_values)
            delta = max(delta, abs(new_V[s] - V[s]))

        V = new_V
        iteration += 1

        if delta < theta:
            break

    return V
```

```python
# ------------------------------------------------
# ✅ Task 3: POLICY EXTRACTION
# ------------------------------------------------
def extract_policy(V):
    Pi = {}

    for s in states:
        if s in terminal_states:
            Pi[s] = None
            continue

        best_a, best_q = None, -999
        for a in ACTIONS:
            q_sa = 0
            for prob, next_s in get_next_states(s, a):
                q_sa += prob * (get_reward(next_s) + gamma * V[next_s])
            if q_sa > best_q:
                best_q = q_sa
                best_a = a
        Pi[s] = best_a

    return Pi
```

```python
# ------------------------------------------------
# ✅ Task 4: VISUALIZATION
# ------------------------------------------------
def print_value_table(V):
    grid = np.zeros((ROWS, COLS))
    for r in range(ROWS):
        for c in range(COLS):
            if (r, c) in walls:
                grid[r][c] = None
            else:
                grid[r][c] = V[(r, c)]
    sns.heatmap(grid, annot=True, fmt=".2f", cmap="coolwarm")
    plt.title("Value Table Heatmap")
    plt.show()

def print_policy_table(Pi):
    symbols = {'up': '^', 'down': 'v', 'left': '<', 'right': '>'}
    grid = [["" for _ in range(COLS)] for _ in range(ROWS)]

    for r in range(ROWS):
        for c in range(COLS):
            if (r, c) in walls:
                grid[r][c] = "#"
            elif (r, c) in terminal_states:
                if (r, c) == (0, 3): grid[r][c] = "+1"
                else: grid[r][c] = "-1"
            else:
                action = Pi[(r, c)]
                grid[r][c] = symbols[action]

    for row in grid:
        print(row)

# ------------------------------------------------
# ✅ Run Everything (Default living penalty = -0.04)
# ------------------------------------------------
print("Running Value Iteration...")
V = value_iteration()
Pi = extract_policy(V)

print("\n✅ Final Policy:")
print_policy_table(Pi)

print("\n✅ Final Value Function Heatmap:")
print_value_table(V)
```
